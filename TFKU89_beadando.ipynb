```sql
SELECT sor_tipus,
    COUNT(*) sorok_szam
FROM 
    (SELECT 
        CASE WHEN LEIRAS IS NULL THEN 'Null'
            WHEN LEN(LEIRAS) = 0 THEN 'Üres'
            ELSE 'Nem Üres' 
            END AS sor_tipus
        FROM Termek) rows
GROUP BY sor_tipus
```

```sql
SELECT MIN(KESZLET) AS 'minimum', 
       MAX(KESZLET) AS 'maximum', 
       AVG(KESZLET) AS 'átlag'
FROM Termek
WHERE KESZLET IS NOT NULL
```

```sql
SELECT 
    MEGYS AS 'típus', 
    COUNT(*) 'sorok_száma'
from Termek
WHERE MEGYS IS NOT null
GROUP BY MEGYS
ORDER BY MEGYS
```
