```sql
SELECT szoba_id,FEROHELY+POTAGY as 'teljes',
case 
when klimas='i' then 'klímás'
when klimas='n' then 'nem klímás'
end as 'Klímás-e'
FROM Szoba

where ferohely+potagy>=3
order by FEROHELY+POTAGY
```

```sql
---1.feladat---
SELECT v.NEV,count(f.FOGLALAS_PK) as 'Foglalások száma'
FROM Vendeg v join Foglalas f on v.USERNEV=f.UGYFEL_FK
where v.NEV not like '%k%'
group by v.nev 
having count(FOGLALAS_PK)>=2
```

```sql
---2.feladat---
SELECT distinct v.NEV
FROM Szallashely sza join szoba sz on sza.SZALLAS_ID=sz.SZALLAS_FK
                        join Foglalas f on sz.SZOBA_ID=f.SZOBA_FK
                        join Vendeg v on f.UGYFEL_FK=v.USERNEV
where sza.HELY like '%Balaton%' and month(f.METTOL) in (12,1,2)
```
